<h3>Step 3</h3>

```python
word_list = ['ardvark', 'baboon','camel']
chosen_word = random.choice(word_list)
print(chosen_word)


display = []
for i in range(0,len(chosen_word)):
    display.append('_')
print(display)

end_of_game = False
while end_of_game == False:
    guess = input('enter any letter\n').lower()
    for i in range(0,len(chosen_word)):
        if chosen_word[i] == guess:
            display[i] = chosen_word[i]
        if '_' not in display:
            end_of_game = True
    print(display)
print('you win')
```

```python
stages = ['''
  +---+
  |   |
  O   |
 /|\  |
 / \  |
      |
=========
''', '''
  +---+
  |   |
  O   |
 /|\  |
 /    |
      |
=========
''', '''
  +---+
  |   |
  O   |
 /|\  |
      |
      |
=========
''', '''
  +---+
  |   |
  O   |
 /|   |
      |
      |
=========''', '''
  +---+
  |   |
  O   |
  |   |
      |
      |
=========
''', '''
  +---+
  |   |
  O   |
      |
      |
      |
=========
''', '''
  +---+
  |   |
      |
      |
      |
      |
=========
''']
word_list = ['ardvark', 'baboon','camel']
chosen_word = random.choice(word_list)
print(chosen_word)


display = []
for i in range(0,len(chosen_word)):
    display.append('_')
print(display)

lives = 6
end_of_game = False
while end_of_game == False:
    guess = input('enter any letter\n').lower()
    for i in range(0,len(chosen_word)):
        if chosen_word[i] == guess:
            display[i] = chosen_word[i]
            print(display)
    if guess not in chosen_word:
        lives -= 1
        print(stages[lives])
        if lives == 0:
            end_of_game =True
            print('You Lose')
    if '_' not in display:
        end_of_game = True
        print('you win')
```
